```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
size = 1000
cluster_1 = pd.DataFrame(dict(
    x=np.random.normal(1, 0.2, size),
    y=np.random.normal(1, 0.2, size))
    )
fig = px.scatter(cluster_1, x='x', y='y', template='plotly_dark')
fig.update_layout(width=700, height=700) # Formatting plot
fig.write_image("cluster_1.png")
fig.show()
```
